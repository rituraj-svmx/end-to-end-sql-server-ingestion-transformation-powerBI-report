## Transformation for all the tables 

```python
table_name = []

for i in dbutils.fs.ls("abfss://bronze@datalakestorageaccontrr.dfs.core.windows.net/Sales/"):
  table_name.append(i.name.split("/")[0])
```

```python
from pyspark.sql.functions import col,to_date

for i in table_name:
    path = "abfss://bronze@datalakestorageaccontrr.dfs.core.windows.net/Sales/" + i + "/"
    df = spark.read.format("parquet").load(path)
    column = df.columns
    
    for col in column:
        if "Date" in col or "date" in col:
            df = df.withColumn(col, to_date(col))

    output_path = "abfss://silver@datalakestorageaccontrr.dfs.core.windows.net/Sales/" + i + "/"
    df.write.format("delta").mode("overwrite").save(output_path)
```
