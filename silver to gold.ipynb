```python
import re

def to_snake(name):
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name)
```

## Changing columns names for all the tables

```python
table_name = []

for i in dbutils.fs.ls("abfss://silver@datalakestorageaccontrr.dfs.core.windows.net/Sales/"):
  table_name.append(i.name.split("/")[0])
```

```python
display(table_name)
```

```python
for i in table_name:
    path = "abfss://silver@datalakestorageaccontrr.dfs.core.windows.net/Sales/" + i + "/"
    df = spark.read.format("delta").load(path)
    
    column_names = df.columns

    for old_col_name in column_names:
        new_col_name = to_snake(old_col_name)
        df = df.withColumnRenamed(old_col_name,new_col_name)

    gold_path = "abfss://gold@datalakestorageaccontrr.dfs.core.windows.net/Sales/" + i + "/"

    df.write.format("delta").mode("overwrite").save(gold_path)
```
